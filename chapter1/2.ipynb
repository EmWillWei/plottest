```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm as cm
```

```python
x = np.linspace(0.5, 3.5, 100)
y = np.sin(x)
y1 = np.random.randn(100)
```

```python
plt.scatter(x,y1, c="0.25", label="scatter figure")
plt.plot(x,y,ls='--',lw=2, label="plot figure")
plt.xlim(0.0, 4.0)
plt.ylim(-3, 3)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.grid(True, ls=':', color='r')
plt.axhline(y=0, c='r', ls='--', lw=2)
plt.axvspan(xmin=1.0, xmax=2.0, alpha=.3, facecolor="y")
plt.annotate("maximum", xy=(np.pi /2, 1.0),xytext=(np.pi/2 + 0.15, 1.5), weight="bold", color="r", arrowprops=dict(arrowstyle='->'))
plt.annotate("spines", xy=(.75, -3), xytext=(.35, -2.25), color='r', arrowprops=dict(arrowstyle='->'))
plt.annotate("", xy=(.2, -3), xytext=(.35, -2.25), color='r', arrowprops=dict(arrowstyle='->'))
```
